```python
import numpy as np
```

```python
#Gradient descent is the process used to get m and b values of the linear regression function so that the line plotted with reg.coef_ and reg_intercept will be as close as possible to the datapoints
#Gradient descent is nothing but taking an initial value of zero for m and b, we try to find the global minimum of a function
#cost function is nothing but the sum of all mean squared errors
#we get a single line when we plot the linear regression function. That line might or might not coincide with all the data points.
#the line is drawn in such a way that the sum of mean squared errors is the minimum
#mean squared error means -> Let 10 be the number of data points for a dataset. let point a be on the linear regression function line, point b be a data point. now (a-b) is squared.
#similarly for every data point, we find the difference between point on the line and datapoint and square it.
#we add all those squared numbers and divide it by the number of data points (Essentially taking the mean of squared errors)


#take random number of iterations and modify based on the result set
#take random learning rate (Start with 0.0001, then 0.001, then 0.01...) as long as the cost function is reducing after each iteration


def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        m_derivative = -(2/n)*sum(x*(y-y_predicted))
        b_derivative = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative
        print(f'iteration = {i} : m_curr = {m_curr}, b_curr = {b_curr}, cost = {cost}')
```

```python
# we are taking the below x and y values based on the sample equation y = 2x + 3 (here m = 2 and b = 3)
#hence we need to keep modifying the learning rate and iteration count till we get m_curr = 2 and b_curr = 3
#the whole point of gradient descent is to find the global minima (that is, minimum possible value of cost for a given function)
#if you observe that the cost function is not reducing anymore is almost stable after many iterations, we can consider it to be the minima
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
